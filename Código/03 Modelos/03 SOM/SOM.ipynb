# Self-Organizing Maps (SOM)

```python
# Biblioteca para a implementação do SOM
from minisom import MiniSom

# Bibliotecas de manipulação de dados
import pandas as pd
import numpy as np

# Biblioteca para a geração de gráficos
import matplotlib.pyplot as plt

# Bblioteca para desligar os avisos do Python
import warnings

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
# Carrega o dataset
df = pd.read_csv(r'..\..\datasets\df_norm_most_common_ncm.csv', encoding='latin-1')
df.info()

# Cria uma cópia da base original
df_subset = df.iloc[:,:].values

# Encontra o tamanho da base
input_size = df_subset.shape[1]

# Define a porcentagem dos top outliers, neste caso 1%. Valor de contaminação
outliers_percentage = 0.01

# Cria um modelo do SOM com um mapa de tamanho 40x40
som = MiniSom(40, 40, df_subset.shape[1], sigma=3, learning_rate=0.5,
              neighborhood_function='triangle', random_seed=26)

# Treina o modelo
som.train(df_subset, 1000000, random_order=False, verbose=True)

# Encontra o Anomaly Score para cada registro
df['anomaly_score'] = np.linalg.norm(som.quantization(df_subset) - df_subset, axis=1)

# Encontra o valor que separa os valores normais das anomalias. Considerando 1% de contaminação
error_treshold = np.percentile(df['anomaly_score'], 100 * (1 - outliers_percentage))

print('Error treshold:', error_treshold)

# Cria uma lista com os top 1% outliers
df_outliers = df[df['anomaly_score'] > error_treshold]
```

```python
# Desenha um u-matrix para encontrar a topologia dos dados.
#  !!! Não tem aplicação prática para este projeto !!!
u_matrix = som.distance_map()
plt.figure(figsize=(15, 12))
plt.pcolor(u_matrix, cmap='inferno_r')
plt.colorbar()
plt.title('U-Matrix')
```

```python
# Ordena a lsita de outliers de forma que o registro com maior Anomaly Score fique no topo
df_outliers['idx'] = df_outliers.index
df_outliers = df_outliers.sort_values(by='anomaly_score', ascending=False)

# Salva os resultados em um CSV
df_outliers.to_csv('output\outliers.csv', index=False)
df_outliers
```
